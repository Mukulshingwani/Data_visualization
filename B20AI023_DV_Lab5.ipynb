#### using SVM classifer

```python
from sklearn.svm import SVC

# Extract the features and target variable
X = df[['SepalLengthCm', 'SepalWidthCm']]
y = df['Species']

# Convert the target variable to numerical values
y = np.where(y=='Iris-setosa', 0, np.where(y=='versicolor', 1, 2))

# Train a linear SVM on the Iris dataset
clf = SVC(kernel='linear')
clf.fit(X, y)

# Define a function to plot the decision boundary
def plot_decision_boundary(clf, X, y):
    x_min, x_max = X['SepalLengthCm'].min() - 0.5, X['SepalLengthCm'].max() + 0.5
    y_min, y_max = X['SepalWidthCm'].min() - 0.5, X['SepalWidthCm'].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02), np.arange(y_min, y_max, 0.02))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.4)
    plt.scatter(X['SepalLengthCm'], X['SepalWidthCm'], c=y, cmap='viridis')

# Plot the decision boundary
plot_decision_boundary(clf, X, y)
plt.title("Linear SVM Decision Boundary")

# Show the plot
plt.show()
```

```python
for col in df2.columns:
    count = df2[col].value_counts(dropna=False).get('?')
    if count is not None:
        print(f"Column '{col}' has {count} missing values represented as '?'.")
    else:
        print(f"Column '{col}' has no missing values represented as '?'.")
```

```python
df2 = df2.replace('?', np.nan)
df2 = df2.dropna()
```

Note: we replace '?' with NaN values and dropped those rows

```python
for col in ['workclass','occupation','native-country']:
    df2[col].fillna(df2[col].mode()[0],inplace=True)

df2.head()
```
